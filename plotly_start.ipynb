```python
df = pd.read_excel('/meantorus.xlsx', sheet_name = 'data')
df.head()
```

```python
df.rename(columns = {'Unnamed: 0':'Регион'}, inplace = True)
df.head()
```

```python
df_plt = pd.melt(df, id_vars = ['Регион'])
df_plt
```

```python
px.line(df_plt, y = 'value', x = 'variable', color = 'Регион', title = 'Динамика ЗП в регионах')
```

```python
px.box(df, y = 'декабрь', points = 'all', hover_name = 'Регион')
```

```python
px.box(df_plt, y = 'value', x = 'variable', hover_name = 'Регион', color = 'variable').update_layout(showlegend = False)
```

```python
df = pd.read_excel('/Эконом_показатели_по_городам.xlsx')
df.head()
```

```python
df.rename(columns = {'Unnamed: 0': 'city'}, inplace = True)
df.head()
```

```python
px.scatter(df, x = 'mean_size_house', y = 'mean_wage', hover_name = 'city', color = 'city').update_layout(showlegend = False)
```

```python
px.scatter(df, x = 'mean_size_house', y = 'mean_wage', hover_name = 'city', color = 'city', size = 'n_pop',
           size_max = 50).update_layout(showlegend = False)
```

```python
bubble_plot = px.scatter(df, x = 'mean_size_house', y = 'mean_wage', hover_name = 'n_pop', color = 'n_pop', size = 'n_pop',
           size_max = 50).update_layout(showlegend = False)
bubble_plot
```

```python
bubble_plot.write_html('/wijiiji.html')
```
